```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
location = []
company_name = []
intern_role = []
intern_dur = []
stipend = []

base_url = "https://internshala.com/internships/software%20development-internship/page-"

for page in range(1,9):
    #print(base_url+str(page))
    r = requests.get(base_url+str(page))
    c = r.content    
    soup = BeautifulSoup(c,'html.parser')
    all = soup.find_all('div', {'class':'internship_meta'})
    for i in range(len(all)):
        intern_role.append(all[i].find_all('a',{'class':'view_detail_button'})[0].text)
    for i in range(len(all)):
        location.append(all[i].find_all('a',{'class':'location_link view_detail_button'})[0].text)
    for i in range(len(all)):
        x = all[i].find_all('div',{'class':'item_body'})[1].text
        x = x.strip('\n')
        x = x.strip(' ')
        intern_dur.append(x)
    for i in range(len(all)):
        stipend.append(all[i].find_all('span',{'class':'stipend'})[0].text)
    for i in range(len(all)):
        y = all[i].find_all('a',class_= 'link_display_like_text view_detail_button')[0].text
        y = y.strip('\n')
        y = y.strip(' ')
        company_name.append(y)
```

```python
dict = {'Intern_Role':intern_role,'Company_Name':company_name,'Location':location,'Duration':intern_dur,'Stipend':stipend}
df = pd.DataFrame(dict)
df
```

```python
df.to_csv("SD_internships_internshaala.csv")
```
